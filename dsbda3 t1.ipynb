```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bank_data_set.csv')
df.head()
```

```python
df.dtypes
```

## countplot: shows frequency of categorical values

```python
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'deposit']
inference_values = ['management', 'married', 'secondary', 'no', 'no', 'no', 'cellular', 'may', 'unknown', 'no']

for i in range(len(categorical_features)):  # Iterate using index
    feature = categorical_features[i]
    inference = inference_values[i]
    
    sns.countplot(x=feature, data=df, order=df[feature].value_counts().index)  # Sorted for better visualization
    plt.xticks(rotation=45)
    plt.title(f'Distribution of {feature}')
    plt.show()
    print(f'Inference: The category with the highest frequency for {feature} is {inference}.')
```

## Barplot: between categorical(job,education,default,housing,loan) and numerical data(balance) 

```python
categorical_features = ['job','education', 'default', 'housing', 'loan','poutcome','deposit']
inference_values = ['retired', 'tertiary', 'no', 'no', 'no', 'other', 'yes']

for i in range(len(categorical_features)):
    feature = categorical_features[i]
    inference = inference_values[i]
    
    sns.barplot(x=feature,y='balance',data=df)
    plt.xticks(rotation=45)
    plt.show()
    print(f'Inference: The category with the highest balance for {feature} is {inference}.')
```

## scatterplot: two numerical columns duration,balance changing hue for job and education

```python
sns.scatterplot(data=df,x='duration',y='balance',hue='education')
plt.show()
print('inference: mostly duration is lower and lower balance but higher duration is shown by education category secondary')
```

```python
sns.scatterplot(data=df,x='previous',y='pdays',hue='contact')
plt.show()
print('inference: previous contacts 0-10 days are made last contacted is also higher till 800 and using cellphone mostly contact made')
```

## pie chart: categorical variables(job,marital,education,default,housing,loan,deposit) distribution in %

```python
categories = ['job','marital','education','default','housing','loan','deposit']
inference_values = ['management','married','secondary','no','no','no','no']

for i in range(len(categories)):
    category = categories[i]
    inference = inference_values[i]
    
    val = df[category].value_counts()
    plt.pie(val,labels=val.index,autopct='%.2f')
    plt.title(f'{category} piechart')
    plt.show()
    print(f'Inference: for feature {category} highest percentage is for {inference}')
```

## Histplot

```python
sns.histplot(data=df,x='duration')
plt.show()
print('inference: shows distribution of duration from 0-500(seconds) being the highest count meaning duration for contact mostly as 500 seconds which 15 minutes')
```

```python
sns.histplot(data=df,x='age')
plt.show()
print('inference: shows distribution of age column from 25-50 ages while highest count from 30-40 age')
```

## Lineplot

```python
sns.lineplot(data=df,x='month',y='campaign')
plt.show()
print('inference: Lineplot: It shows relationship between Month and campaign which month number of campaigns in months jun,jul,aug most campaigns')
```

```python
sns.lineplot(data=df,x='day',y='campaign')
plt.show()
print('inference: Lineplot: It shows relationship between day and campaign on which day number of campaigns more campaigns 25-30 days')
```

## Pairplot

```python
sns.pairplot(df)
plt.show()
print('inference: relationships between variables within a data set like age,duration,balance,day,campaign,pdays,previous')
```

## Box Plot

```python
sns.boxplot(data=df,x='job',y='balance')
plt.xticks(rotation=45)
plt.show()
print('inference:Box Plot: comparisons(distribution) for job balance is less in mostly all types and mostly outliers are shown for management,retired,entrepreneur,etc meaning among these categoies some of them have high balance(yearly income)')
```

```python
sns.boxplot(data=df,y='campaign',x='poutcome')
plt.show()
print('inference:Box Plot: comparisons(distribution) unknown outcome of campaign has highest outliers mostly outcomes of previous known campaigns are which also has highest cotacts')
```
