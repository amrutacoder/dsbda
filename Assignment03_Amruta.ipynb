```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
dt=pd.read_csv('cancer_data_set (1).csv' )
dt
```

```python
dt.isnull().sum()
```

```python
n=dt.select_dtypes(include='number')
m=n.mean()
dt=dt.fillna(m)
dt
```

```python
dt.isnull().sum()
```

```python
# Bar plot 
plt.figure(figsize=(6, 4))
sns.barplot(x=dt["diagnosis"].value_counts().index, y=dt["diagnosis"].value_counts().values)
plt.title("Bar Plot of Diagnosis Count")
plt.xlabel("Diagnosis (B: Benign, M: Malignant)")
plt.ylabel("Count")
plt.show()

#The bar plot shows that the number of benign (B) cases is higher than malignant (M) cases in the dataset. B-357 M-212
```

```python
# Count plot 
plt.figure(figsize=(6, 4))
sns.countplot(x=dt["diagnosis"])
plt.title("Diagnosis Distribution")
plt.xlabel("Diagnosis (B: Benign, M: Malignant)")
plt.ylabel("Count")
plt.show()
#The count plot shows that the number of benign (B) cases is greater than malignant (M) cases in the dataset.
```

```python
# Scatter plot of radius_mean vs texture_mean colored by diagnosis
plt.figure(figsize=(8, 5))
sns.scatterplot(x=dt["radius_mean"], y=dt["texture_mean"], hue=dt["diagnosis"], palette=["#ff9999", "#66b3ff"])
plt.title("Scatter Plot of Radius Mean vs Texture Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.legend(title="Diagnosis")
plt.show()
#Malignant tumors usually have larger size and higher radius_mean and texture_mean, showing a clear separation from benign tumors.
```

```python
features = ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "diagnosis"]
sns.pairplot(dt[features], hue="diagnosis")
plt.show()

#Malignant tumors generally have higher values for features like radius_mean, perimeter_mean, and area_mean, making them distinguishable from benign tumors.
```

```python
# Pie chart for diagnosis distribution
dt["diagnosis"].value_counts().plot.pie(autopct="%1.1f%%", colors=["#66b3ff", "#ff9999"])  
plt.title("Diagnosis Distribution")  
plt.show()

#More benign cases than malignant cases.
```

```python
# Histogram for radius_mean
sns.histplot(dt["radius_mean"], bins=30, kde=True, color="skyblue")  
plt.title("Distribution of Radius Mean")  
plt.xlabel("Radius Mean")  
plt.ylabel("Count")  
plt.show()

#Most tumors have a similar radius_mean size.
#Some tumors are much larger or smaller than the average.
```

```python
plt.figure(figsize=(8, 5))  
sns.lineplot(x=dt.index, y=dt["radius_mean"], hue=dt["diagnosis"])  
plt.title("Line Plot of Radius Mean")  
plt.xlabel("Index")  
plt.ylabel("Radius Mean")  
plt.show()
#The radius_mean values change across different samples showing that tumor sizes are not consistent.
# Some tumors have a much larger radius than others
```
